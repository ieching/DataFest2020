```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
```

```python
cases_df = pd.read_csv("data/us-cases.csv")

aal_df = pd.read_csv("data/AAL.csv")
dal_df = pd.read_csv("data/DAL.csv")
luv_df = pd.read_csv("data/LUV.csv")
ual_df = pd.read_csv("data/UAL.csv")

flights_df = pd.read_csv("data/number-of-commercial-fli.csv")

tw1 = pd.read_csv("data/coronavirus-covid19-tweets/2020-03-00 Coronavirus Tweets (pre 2020-03-12).CSV")
tw2 = pd.read_csv("data/coronavirus-covid19-tweets/2020-03-12 Coronavirus Tweets.CSV")
tw3 = pd.read_csv("data/coronavirus-covid19-tweets/2020-03-13 Coronavirus Tweets.CSV")
tw4 = pd.read_csv("data/coronavirus-covid19-tweets/2020-03-14 Coronavirus Tweets.CSV")
tw5 = pd.read_csv("data/coronavirus-covid19-tweets/2020-03-15 Coronavirus Tweets.CSV")
tw6 = pd.read_csv("data/coronavirus-covid19-tweets/2020-03-16 Coronavirus Tweets.CSV")
tw7 = pd.read_csv("data/coronavirus-covid19-tweets/2020-03-17 Coronavirus Tweets.CSV")
tw8 = pd.read_csv("data/coronavirus-covid19-tweets/2020-03-18 Coronavirus Tweets.CSV")
tw9 = pd.read_csv("data/coronavirus-covid19-tweets/2020-03-19 Coronavirus Tweets.CSV")

tw = [tw1, tw2, tw3, tw4, tw5, tw6, tw7, tw8, tw9]
tweets_df = pd.concat(tw, axis=0)

hashtag_df = pd.read_csv("data/coronavirus-covid19-tweets/Hashtags.CSV")
```

```python
tweets_df.info()
```

```python
flights_df.columns = ['TimeStamp', 'Average', 'NumberOfFlights']
flights_df.head()
```

```python
aal_df.drop(columns=['High', 'Low', 'Close', 'Adj Close'], inplace=True)
dal_df.drop(columns=['High', 'Low', 'Close', 'Adj Close'], inplace=True)
luv_df.drop(columns=['High', 'Low', 'Close', 'Adj Close'], inplace=True)
ual_df.drop(columns=['High', 'Low', 'Close', 'Adj Close'], inplace=True)

aal_df.columns = ['TimeStamp', 'AALOpen', 'AALVolume']
dal_df.columns = ['TimeStamp', 'DALOpen', 'DALVolume']
luv_df.columns = ['TimeStamp', 'LUVOpen', 'LUVVolume']
ual_df.columns = ['TimeStamp', 'UALOpen', 'UALVolume']

stock_df = aal_df.merge(dal_df, how="inner", on="TimeStamp")
stock_df = stock_df.merge(luv_df, how="inner", on="TimeStamp")
stock_df = stock_df.merge(ual_df, how="inner", on="TimeStamp")

stock_df.head()
```

```python
cases_df.columns = ['TimeStamp', 'Cases', 'Deaths']
cases_df.head()
```

```python
merged_df = cases_df.merge(flights_df, how="inner", on="TimeStamp")
merged_df = merged_df.merge(stock_df, how="inner", on="TimeStamp")
merged_df.head()
```

```python
merged_df.to_csv("data/processed/unscaled_cases+stocks+flights.csv", index=False)
```

```python
time = merged_df['TimeStamp']
to_scale = merged_df.drop(columns=['TimeStamp'])
columns = to_scale.columns

scaler = MinMaxScaler()
scaled = scaler.fit_transform(to_scale)
scaled = pd.DataFrame(scaled, columns=columns)

final_df = pd.concat([time, scaled], axis=1)
final_df.head()
```

```python
final_df.to_csv('data/processed/scaled_cases+stocks+flights.csv', index=False)
```

## Twitter Text Analysis

```python
tweets_df.head()
```

```python
hashtag_df.head()
```

```python
def lower_text(text):
    try:
        return text.lower()
    except:
        return 'failed'
            
hashtag_df['hashtag'] = hashtag_df['hashtag'].apply(lower_text)
```

```python
keywords = ['air','airplane','aeroplane','airways','aircraft','airlines','plane','flight','flights','fly','flying',
            'airport','lax','customs','travel','travels','traveling','traveler']

filtered_hashtag_df = hashtag_df[hashtag_df['hashtag'].isin(keywords)]

twitter_df = tweets_df.merge(filtered_hashtag_df, on='status_id', how="inner")
twitter_df.info()
```

```python
twitter_df.head()
```

```python
twitter_df.to_csv("data/processed/airlines_twitter.csv", index=False)
```
